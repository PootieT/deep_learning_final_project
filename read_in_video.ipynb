```python
def read_range(line, idx):
    """
    helper function to read the ranges of frames in text file
    """
    lb = idx
    while line[lb] != ' ':
        lb -= 1
    lb += 1
    
    ub = idx
    while ub != len(line) and line[ub] != ',':
        if ub == len(line):
            break
        ub += 1
    return list([int(line[lb:idx]), int(line[idx+1:ub])])

#change to your own directory
os.chdir("C:/serious stuff/Rice/Senior Fall/ELEC 576 (Deep Learning)/Final Project/deep_learning_final_project/")

file_object = open("valid_frames.txt",'r')
valid_frame_dict = {}
for line in file_object.readlines(0):
    idx = line.find("frames")
    newline = line[0:idx-2] + ", " + line[idx+7:]
    idx_list = [pos for pos, char in enumerate(newline) if char == '-']
    frame_list = list([read_range(newline, idx_list[0]), read_range(newline, idx_list[1]),read_range(newline, idx_list[2]),read_range(newline, idx_list[3])])
    valid_frame_dict[newline[0:idx-2]] = frame_list

```
